```python
x = df[['sqft_living','sqft_lot','sqft_above','sqft_basement']]
y = df['price']
```

```python
# Adding a column of ones to X for the intercept term
x = np.append(np.ones((x.shape[0],1)),x,axis=1)
```

```python
def cost_function(x,y,theta):
    m = len(y) # number of samples
    predictions = x.dot(theta) #predictions of the model
    errors = predictions - y #errors of the predictions
    cost = (1/(2*m)) * np.sum(errors**2) # mean squared error
    return cost
```

```python
def gradient_descent(x,y,theta,alpha,iterations):
    m = y.size # number of samples
    theta = np.zeros((x.shape[1],1)) # initializing theta with zeroes
    cost_history = []
    
    for i in range(iterations):
        predictions = np.dot(x,theta)
        
        cost = (1/(2*m))*np.sum(np.square(predictions-y))
        d_theta = (1/m)*np.dot(x.T, predictions-y)
        theta = theta - alpha * d_theta # updating theta values
        cost_history.append(cost)
    return theta, cost_history
```

```python
# setting the learning rate and number of iterations
alpha = 0.001
iterations = 100
```

```python
# performing gradient descent and getting final theta and cost history
theta, cost_history = gradient_descent(x,y,theta,alpha,iterations)
```
